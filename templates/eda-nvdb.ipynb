```python
## Data manipulation
import pandas as pd
import geopandas

# Options for pandas
pd.options.display.max_columns = 50
pd.options.display.max_rows = 30
```

```python
df = pd.read_csv(config.get_processed_file("etl_e16.csv"), low_memory=False)
```

```python
gs = geopandas.GeoSeries.from_wkt(df["geometri"])
gdf = geopandas.GeoDataFrame(df, geometry=gs, crs="EPSG:5973")
```
